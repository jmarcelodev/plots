```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('train_test.csv')
```

```python
# Exclude certain columns from the DataFrame
columns_to_exclude = ['fid', 'lat', 'long', 'talhao', 'wkt_geom', 'Altitude1', 'frutos_ver', 'total_frut']
matrix = df1.drop(columns=columns_to_exclude)

# Plot the heatmap
fig, axes = plt.subplots(figsize=(12, 8))
sns.heatmap(matrix.corr(), cmap='viridis', annot=True, fmt=".2f")
plt.title('Correlao das Variveis')
plt.show()
```
